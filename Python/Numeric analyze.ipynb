```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
belief = np.random.rand(10)
```

```python
# 0 location, 1 scene, 2 meet_character, 3 hold_object, 4 interact 5 pickUp/Drop 6 eat 7 health 8 energy 9 wealth
names = ["location", "scene" ,"meet_character" ,"hold_object" ,"interact" ,"pickUp/Drop" ,"use" ,"health" ,"energy" ,"wealth"]
```

```python
records = []
for i in range(1000):
    records.append([_ for _ in belief])
    idx = np.random.randint(10)
    if(idx == 3 or idx == 7 or idx == 5 or idx == 6):
        belief[idx] = min(belief[idx]*1.05, 1)
    elif(idx == 2):
        pass
    else:
        belief[idx] *= 0.95
    
```

```python
records = np.asarray(records)
```

```python
for i in range(len(records[0])):
    plt.plot(records[:,i],label = names[i])
plt.legend()

plt.show()
```
